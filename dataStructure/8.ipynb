```python
class Position:
    _data:int
    _next:'Position' = None
    _prev:'Position' = None
    def __init__(self, data:int):
        self._data = data
    
    def element(self):
        return self._data


class PositionalList:
    _first:'Position' = None
    _last:'Position' = None
    _len:int = 0
    def first(self):
        return self._first

    def last(self):
        return self._last

    def before(self, p:'Position'):
        return p._prev

    def after(self, p:'Position'):
        return p._next

    def is_empty(self):
        if self._first == None and self._last == self._first:
            return True
        return False

    def __len__(self):
        return self._len

    def __iter__(self):
        cur = self.first()
        while(cur):
            yield cur
            cur = self.after(cur)
    
    def add_first(self, e:int):
        p = Position(e)
        self._len += 1
        if self._first:
            temp = self._first
            self._first = p
            p._next = temp
            temp._prev = p
            return p
        self._first = p
        self._last = p
        return p  

    def add_last(self, e:int):
        p = Position(e)
        self._len += 1
        if self._last:
            temp = self._last
            self._last = p
            p._prev = temp
            temp._next = p
            return p
        self._first = p
        self._last = p
        return p
    
    def add_before(self, p:'Position', e:int):
        cur = Position(e)
        self._len += 1
        if p._prev:
            temp = p._prev
            temp._next = cur
            p._prev = cur
            cur._prev = temp
            cur._next = p
            return cur
        
        self._first = cur
        cur._next = p
        p._prev = cur
        return cur

    def add_after(self, p:'Position', e:int):
        cur = Position(e)
        self._len += 1
        if p._next:
            temp = p._next
            temp._prev = cur
            p._next = cur
            cur._next = temp
            cur._prev = p
            return cur
        
        self._last = cur
        cur._prev = p
        p._next = cur
        return cur

    def delete(self, p:'Position'):
        if p == self.first():
            temp = p._next
            p._next = None
            self._first = temp
            temp._prev = None
        elif p == self.last():
            temp = p._prev
            p._prev = None
            self._last = temp
            temp._next = None
        else:
            p._prev._next = p._next
            p._next._prev = p._prev
            p._prev = None
            p._next = None
        self._len -= 1

L = PositionalList()
L.add_first(10)
t = L.add_last(100)
L.add_before(t, 90)
L.add_before(t, 80)
L.delete(t)
L.delete(L.first())
for i in L:
    print(i.element())
```
